```python
class RangeExeptionError(Exception):
    def __init__(self, message="please enter a range between 1-4\n"):
        self.message = message
        super().__init__(self, self.message)

class Calculator:
    def __init__(self):
        self._result = None
        
        while True:
            try:
                self.num1, self.num2 = map(int, input("Please enter 2 numbers: ex:(10, 20) ").split(","))
                
            except Exception as e:
                print(f"{e.__class__.__name__}: {e.message if hasattr(e, 'message') else e.__str__()}\n")
                continue
            else:
                break
            
    @property
    def result(self):
        return self._result

class Sum(Calculator):
    def __init__(self):
        super().__init__()
    
    @property
    def result(self):
        self._result = self.num1 + self.num2
        
        return self._result

class Sub(Calculator): 
    def __init__(self):
        super().__init__()
        
    @property
    def result(self):
        self._result = self.num1 - self.num2
        
        return self._result
        
class Multiply(Calculator):
    def __init__(self):
        super().__init__()
        
    @property
    def result(self):
        self._result = self.num1 * self.num2
        
        return self._result
    
class Division(Calculator): 
    def __init__(self):
        super().__init__()
        
    @property
    def result(self):
        try:
            self._result = self.num1 / self.num2
        except ZeroDivisionError as e:
            print(f"{e.__class__.__name__}: {e.message if hasattr(e, 'message') else e.__str__()}\n")
        
        return self._result
    

while True:
    while True:
        try:
            operator = int(input("Enter the operator you want:\n1-add\n2-sub\n3-multiply\n4-division\n"))

            if operator not in range(1,5):
                raise RangeExeptionError
                
        except Exception as e:
            print(f"{e.__class__.__name__}: {e.message if hasattr(e, 'message') else e.__str__()}\n")
            continue
        else:
            break
    
    if operator == 1:
        calculator = Sum()
    elif operator == 2:
        calculator = Sub()
    elif operator == 3:
        calculator = Multiply()
    else:
        calculator = Division()
    
    print(calculator.result)
    
    again = input("Do you want another operation? y for yes ")
    
    if again == "y":
        continue
    else:
        break

print("Thanks.")
```
